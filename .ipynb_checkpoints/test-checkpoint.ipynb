```python
#將入睡，起床時間轉換為時間戳，睡眠時間以時間戳代替。
import time
i = 0
for i in range(len(df1['Start'])):
    try:
        struct_time = time.strptime(df1["Start"][i], "%Y-%m-%d %H:%M:%S")
        time_stamp = int(time.mktime(struct_time))
        df1.loc[i, 'Start'] = time_stamp
    except:
        print("error", i)
        break
j = 0
for j in range(len(df1['End'])):
    try:
        struct_time = time.strptime(df1["End"][j], "%Y-%m-%d %H:%M:%S")
        time_stamp = int(time.mktime(struct_time))
        df1.loc[j, 'End'] = time_stamp
    except:
        print("error", j)
        break
k = 0
for k in range(len(df1['Time in bed'])):
    try:
        df1.loc[k, 'Time in bed'] = df1.loc[k, 'End'] - df1.loc[k, 'Start']
    except:
        print("error", k)
        break
df1['Start']
```

```python
# 將睡眠品質的值轉成數值
i = 0
for i in range(len(data1['Sleep quality'])):
    characters = "%"
    x = 0
    try:
        for x in range(len(characters)):
            data1.loc[i, 'Sleep quality'] = float(data1.loc[i, 'Sleep quality'].replace(characters[x],""))    
    except:
        print("error", i)
        break
data1['Sleep quality']
```

```python
#將資料做數值轉換，並建立一個新column儲存它們
j = 0
for j in range(len(df1['Sleep Notes'])):
    list_Sleep_notes = ["Stressful day", 'Drank coffee', 'Drank tea', 'Ate late', 'Worked out'] #所有活動
    Sleep_notes = df1['Sleep Notes'][j]
    float_types = df1['Sleep Notes'][0]
    amount = 0
    try:
        if (type(Sleep_notes) == type(float_types)):
            amount = 0
        else:
            length = len(Sleep_notes.split(':'))
            if length >=2:
                i = 0
                for i in range(length):
                    if Sleep_notes.split(':')[i] == list_Sleep_notes[0]:
                        amount-=2
                    elif Sleep_notes.split(':')[i] == list_Sleep_notes[1]:
                        amount-=2
                    elif Sleep_notes.split(':')[i] == list_Sleep_notes[2]:
                        amount-=1
                    elif Sleep_notes.split(':')[i] == list_Sleep_notes[3]:
                        amount-=1
                    elif Sleep_notes.split(':')[i] == list_Sleep_notes[4]:
                        amount+=2
            else:
                i = 0
                for i in range(length):
                    if Sleep_notes.split(':')[i] == list_Sleep_notes[0]:
                        amount-=2
                    elif Sleep_notes.split(':')[i] == list_Sleep_notes[1]:
                        amount-=2
                    elif Sleep_notes.split(':')[i] == list_Sleep_notes[2]:
                        amount-=1
                    elif Sleep_notes.split(':')[i] == list_Sleep_notes[3]:
                        amount-=1
                    elif Sleep_notes.split(':')[i] == list_Sleep_notes[4]:
                        amount+=2
    except:
        print("error", j)   
    df1.loc[j, 'Sleep Notes'] = amount
type(df1['Sleep Notes'][0])
```
